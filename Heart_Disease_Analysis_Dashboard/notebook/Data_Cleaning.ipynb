```python
df
```

```python
df['age_grp'] = df['age_grp'].apply(lambda x: 'Child' if x <= 12 
                                    else 'Teen' if 13 <= x <= 19 
                                    else 'Young Adult' if 20 <= x <= 35 
                                    else 'Middle-Aged' if 36 <= x <= 55 
                                    else 'Senior')
```

```python
df['cholesterol_category'] = df['chol'].apply(lambda x: 'Desirable' if x < 200 
                                              else 'Borderline High' if 200 <= x <= 239 
                                              else 'High')
```
